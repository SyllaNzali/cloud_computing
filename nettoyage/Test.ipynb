```python
import pandas as pd

# Charger le fichier CSV
file_path = "../result/clean_data.csv"

df = pd.read_csv(file_path)

print("✅ Dataset loaded successfully\n")

# ✅ 1. Nombre de lignes et colonnes
print("Shape of dataset:", df.shape)
print("-" * 50)

# ✅ 2. Aperçu des données
print("First 5 rows:")
print(df.head())
print("-" * 50)

# ✅ 3. Infos générales
print("Dataset info:")
print(df.info())
print("-" * 50)

# ✅ 4. Valeurs manquantes
print("Missing values per column:")
print(df.isnull().sum())
print("-" * 50)

# ✅ 5. Valeurs uniques (important pour analyse)
print("Unique values:")

print("Drugs:", df["drug_name"].nunique())
print("Adverse events:", df["adverse_event"].nunique())
print("Countries:", df["country"].nunique())
print("-" * 50)

# ✅ 6. Top 10 drugs
print("Top 10 drugs:")
print(df["drug_name"].value_counts().head(10))
print("-" * 50)

# ✅ 7. Top adverse events
print("Top 10 adverse events:")
print(df["adverse_event"].value_counts().head(10))
print("-" * 50)

# ✅ 8. Vérifier les doublons
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

print("\n✅ Verification completed")
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Convert date
df["report_date"] = pd.to_datetime(df["report_date"], format="%Y%m%d", errors="coerce")

# Optional: remove missing values
df = df.dropna()

print("✅ Clean shape:", df.shape)

# Save again
df.to_csv("../result/clean_data1.csv", index=False)
```
